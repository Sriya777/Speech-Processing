```python
data, sr = sf.read(file_path)

bit_depth = data.dtype.itemsize * 8
num_samples = len(data)
duration = num_samples / sr

print("Sampling Rate:", sr, "Hz")
print("Bit Depth:", bit_depth, "bits")
print("Number of Samples:", num_samples)
print("Total Duration:", duration, "seconds")
```

STEP 4: Plot Speech Waveform (Time Domain)

```python
time = np.arange(num_samples) / sr

plt.figure(figsize=(10,4))
plt.plot(time, data)
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("Speech Signal Waveform")
plt.grid()
plt.show()
```

STEP 5: Active Speech Duration (Energy-Based)

```python
energy = data**2
threshold = 0.01 * np.max(energy)

active_samples = np.sum(energy > threshold)
active_duration = active_samples / sr

print("Active Speech Duration:", active_duration, "seconds")
```

STEP 6: Slicing (First 2 Seconds)

```python
slice_samples = int(2 * sr)
sliced_signal = data[:slice_samples]

time_slice = np.arange(len(sliced_signal)) / sr

plt.figure(figsize=(10,4))
plt.plot(time_slice, sliced_signal)
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("First 2 Seconds of Speech")
plt.grid()
plt.show()
```

STEP 7: Normalization

```python
normalized_signal = sliced_signal / np.max(np.abs(sliced_signal))

plt.figure(figsize=(10,4))
plt.plot(time_slice, normalized_signal)
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("Normalized Speech Signal")
plt.grid()
plt.show()
```

STEP 8: Amplification & De-Amplification

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1. Sample signal (replace this with your audio/signal if needed)
# -----------------------------
fs = 44100                      # Sampling frequency
t = np.linspace(0, 2, fs*2)     # 2 seconds
signal = np.sin(2*np.pi*5*t) * np.exp(-t/2)  # example signal

# -----------------------------
# 2. Normalize signal
# -----------------------------
normalized_signal = signal / np.max(np.abs(signal))

# -----------------------------
# 3. Amplification & De-amplification
# -----------------------------
gain_amp = 2
gain_deamp = 0.5

amplified = gain_amp * normalized_signal
de_amplified = gain_deamp * normalized_signal

# -----------------------------
# 5. Print numerical comparison (THIS PROVES DIFFERENCE)
# -----------------------------
def signal_stats(signal, name):
    peak = np.max(np.abs(signal))
    rms = np.sqrt(np.mean(signal**2))
    print(f"{name}")
    print(f"  Peak Amplitude : {peak:.4f}")
    print(f"  RMS Value      : {rms:.4f}")
    print("-" * 35)

signal_stats(normalized_signal, "Original Signal")
signal_stats(amplified, "Amplified Signal (Gain = 2)")
signal_stats(de_amplified, "De-Amplified Signal (Gain = 0.5)")

# -----------------------------
# 6. Print gain ratios
# -----------------------------
print("Gain Ratio (Amplified / Original):",
      np.max(np.abs(amplified)) / np.max(np.abs(normalized_signal)))

print("Gain Ratio (De-Amplified / Original):",
      np.max(np.abs(de_amplified)) / np.max(np.abs(normalized_signal)))
```

STEP 9: Up-Sampling & Down-Sampling

```python
#Up-sampling (Factor = 2)
upsampled = np.repeat(sliced_signal, 2)
time_up = np.arange(len(upsampled)) / (2 * fs)

plt.figure(figsize=(12,4))
plt.plot(time_up, upsampled)
plt.title("Up-sampled Speech Signal (Factor 2)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
#Down-sampling (Factor = 2)
downsampled = sliced_signal[::2]
time_down = np.arange(len(downsampled)) / (fs / 2)

plt.figure(figsize=(12,4))
plt.plot(time_down, downsampled)
plt.title("Down-sampled Speech Signal (Factor 2)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# Up-sampling (factor = 2)
upsampled = signal.resample(normalized_signal, len(normalized_signal)*2)

# Down-sampling (factor = 2)
downsampled = normalized_signal[::2]

plt.figure(figsize=(10,4))
plt.plot(upsampled)
plt.title("Up-Sampled Signal")
plt.grid()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(downsampled)
plt.title("Down-Sampled Signal")
plt.grid()
plt.show()
```

STEP 10: Voiced / Unvoiced / Silence Identification (Visual)

```python
#Energy-based classification
# Select one channel for energy calculation
sliced_signal_mono = sliced_signal[:, 0] # Taking the first channel
energy_slice = sliced_signal_mono**2

silence_th = 0.01 * np.max(energy_slice)
voiced_th = 0.1 * np.max(energy_slice)

labels = np.zeros(len(energy_slice))

labels[energy_slice < silence_th] = 0       # Silence
labels[(energy_slice >= silence_th) & (energy_slice < voiced_th)] = 1  # Unvoiced
labels[energy_slice >= voiced_th] = 2       # Voiced

plt.figure(figsize=(12,4))
# Plot against the selected channel for consistency
plt.scatter(time_slice, sliced_signal_mono, c=labels, cmap="viridis", s=1)
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("Voiced / Unvoiced / Silence Regions")
plt.grid()
plt.show()
```
